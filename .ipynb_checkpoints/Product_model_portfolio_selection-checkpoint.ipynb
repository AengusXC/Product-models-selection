```python
############################################ AHP for finding features weights############################################
print('========================================== Input the AHP features data =======================================')
AHP_data = pd.read_csv('CarFeature_AHP.csv') 
DropColumn = AHP_data.columns.drop(['Features']) 
AHP_data = AHP_data[DropColumn]
print(AHP_data)
AHP_data = np.array(AHP_data, dtype=float)
eigenvalues,eigenvectors = np.linalg.eig(AHP_data)              
max_idx = np.argmax(eigenvalues)
max_eigen = eigenvalues[max_idx].real
eigen = eigenvectors[:, max_idx].real
eigen_weight_bef = eigen / eigen.sum()
print('Maximun eigen value λ is:',max_eigen)
print('Eigen vector W for the maximum eigen value λ is:\n',eigen_weight_bef)
n = len(eigen_weight_bef)
RI = (0, 0, 0.58, 0.9, 1.12, 1.24, 1.32, 1.41, 1.45, 1.49, 1.51, 1.54, 1.56, 1.57, 1.58, 1.59, 1.59, 1.60, 1.60, 1.60) 
CI = (max_eigen - n) / (n - 1)
CR = CI / RI[n-1]
if CR < 0.1:
    print('CR =',CR,'→Result pass consistency check')
    eigen_weight = eigen_weight_bef
else:
     print('Fail to pass consistency check, adjust CarFeature_AHP date ')
```

```python
##############################################Portfolio optimization via Monte Carlo method#########################################

def portfolio_performance(weights, miu, cov_matrix):
    returns = np.sum(miu*weights )
    std = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))) 
    return std, returns

def random_portfolios(num, miu, cov_matrix, rf):
    results = np.zeros((3,num)) 
    weights_record = []
    for i in range(num):
        weights = np.random.rand(len(miu))  
        weights /= np.sum(weights)  
        weights_record.append(weights)
        portfolio_std_dev, portfolio_return = portfolio_performance(weights, miu, cov_matrix)
        results[0,i] = portfolio_std_dev 
        results[1,i] = portfolio_return  
        results[2,i] = (portfolio_return - rf) / portfolio_std_dev #sharpe ratio
    return results, weights_record

def display(miu,var, cov_matrix, num, rf,market_ret, market_vol, n):
    resultMatrix = np.zeros((2,n))
    WeightMatrix = np.zeros((len(miu),n)) # 12 rows n cols   
    for i in range (n):       
        results, weights = random_portfolios(num, miu, cov_matrix, rf) 
        max_sharpe_idx = np.argmax(results[2]) 
        sdp, rp, wp= results[0,max_sharpe_idx], results[1,max_sharpe_idx], weights[max_sharpe_idx] 
        df_1=pd.DataFrame(wp*100,columns=['weight allocation'])
        max_sharpe_allocation=pd.concat([dataset_cols2,df_1],axis=1)
        min_vol_idx = np.argmin(results[0]) 
        sdp_min, rp_min, wp_min = results[0,min_vol_idx], results[1,min_vol_idx], weights[min_vol_idx] 
        df_2=pd.DataFrame(wp_min*100,columns=['weight allocation']) 
        min_vol_allocation=pd.concat([dataset_cols2,df_2],axis=1) 
        print("-"*100)
        print("==================================Maximum Sharpe Ratio Portfolio Allocation==================================\n")
        print("Return of portfolio:", round(rp,2),'%')
        print("Volatility of portfolio:", round(sdp,2))
        print("\n")
        print(df_1)
        print("==================================Minimum Volatility Portfolio Allocation==================================\n")
        print("Return of portfolio:", round(rp_min,2),'%')
        print("Volatility of portfolio:", round(sdp_min,2))
        print("\n")
        print(df_2)
        plt.figure(figsize=(10, 7))
        plt.scatter(results[0,:],results[1,:],c=results[2,:],cmap='YlGnBu', marker='o', s=20, alpha=0.3)
        plt.colorbar()
        plt.scatter(sdp,rp,marker='*',color='r',s=100, label='Maximum Sharpe ratio')
        plt.scatter(sdp_min,rp_min,marker='o',color='g',s=100, label='Minimum volatility', alpha = 1/5)
        plt.scatter(market_vol,market_ret,marker='d',color='b',s=100, label='Market portfolio') 
        plt.text(sdp,rp,(round(sdp,2),round(rp,2)),color='r')
        plt.text(sdp_min,rp_min,(round(sdp_min,2),round(rp_min,2)),color='g')
        plt.title('Simulated Portfolio by Monte Carlo method')
        plt.xlabel('Volatility')
        plt.ylabel('Returns %')
        plt.legend(labelspacing=0.8)               
##############figure saving#############        
        plt.savefig("F:\Python codes\Product model portfolio selection\Result\pic{}.jpg".format(i)) 
        plt.show() 
        plt.clf()       
        resultMatrix[0,i] = rp #return data 
        resultMatrix[1,i] = sdp  # volatility data
        WeightMatrix[:,i] = wp*100    
    label_1=['return','volatility']
    data_result_1 = pd.DataFrame(resultMatrix,index = label_1)
    data_result_1.to_csv("F:\Python codes\Product model portfolio selection\Result\Result_rp_sdp.csv", mode='a+', encoding="utf-8-sig")
    data_result_2 = pd.DataFrame(WeightMatrix)
    data_result_2.to_csv("F:\Python codes\Product model portfolio selection\Result\Result_weight.csv", mode='a+', encoding="utf-8-sig")
```

```python
##############################################Gradient descent method and comparision##############################################
import scipy.optimize 

def portfolio_volatility(weights):
    return portfolio_performance(weights, miu, cov_matrix)[0]

def neg_sharpe_ratio(weights, miu, cov_matrix, rf):
    p_var, p_ret = portfolio_performance(weights, miu, cov_matrix)
    return -(p_ret - rf) / p_var

def max_sharpe_ratio(miu, cov_matrix, rf):
    num_assets = len(miu)
    args = (miu, cov_matrix, rf)
    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
    bound = (0.0,1.0)
    bounds = tuple(bound for asset in range(num_assets))
    result = scipy.optimize.minimize(neg_sharpe_ratio, num_assets*[1./num_assets,], args=args,method='SLSQP', bounds=bounds, constraints=constraints)
    return result

def display_calculated_ef(miu, cov_matrix, market_ret, market_vol,num, rf):    
    results, _ = random_portfolios(num,miu, cov_matrix, rf)
    max_sharpe = max_sharpe_ratio(miu, cov_matrix, rf)
    sdp_m, rp_m = portfolio_performance(max_sharpe['x'], miu, cov_matrix)
    max_sharpe_allocation = pd.DataFrame(max_sharpe.x,columns=['allocation'])
    max_sharpe_allocation.allocation = [round(i*100,3)for i in max_sharpe_allocation.allocation]
    max_sharpe_allocation = max_sharpe_allocation.T
    print ("-"*80)
    print ("Maximum Sharpe Ratio Portfolio Allocation Gradient descent method \n")
    print("Return of portfolio:", round(rp_m,2),'%')
    print("Volatility of portfolio:", round(sdp_m,2))
    print ("\n")
    print (max_sharpe_allocation)
    plt.figure(figsize=(10, 7))
    plt.scatter(sdp_m,rp_m,marker='*',color='r',s=100, label='Maximum Sharpe Ratio Portfolio')  
    x0 = np.array([1.0 / len(miu) for x in range(len(miu))])
    bounds = tuple((0, 1) for x in range(len(miu)))
    given_r = np.arange(market_ret-0.5,market_ret+1, 0.01)
    risk = []              
    for k in given_r:
        constraints = [{'type': 'eq', 'fun': lambda x: sum(x) - 1},
               {'type': 'eq', 'fun': lambda x: sum(x * miu) - k}]
        outcome = scipy.optimize.minimize(portfolio_volatility, x0=x0, constraints=constraints, bounds=bounds)
        risk.append(outcome.fun)
    plt.plot(risk, given_r, linestyle='-.', color='black', label='efficient frontier')  
    constraints = {'type': 'eq', 'fun': lambda x: sum(x) - 1}
    minv = scipy.optimize.minimize(portfolio_volatility, x0=x0, constraints=constraints, bounds=bounds).fun
    minvr = sum(scipy.optimize.minimize(portfolio_volatility, x0=x0, constraints=constraints, bounds=bounds).x * miu)
    plt.scatter(minv, minvr,marker='*',color='b',s=100, label='Global Minimum Variance Portfolio')  
    plt.title('Optimal Portfolio Stimulation by Gradient descent method ')
    plt.xlabel('Volatility')
    plt.ylabel('Returns %')
    plt.legend(labelspacing=0.8)
    plt.figure(figsize=(10, 7))   
    plt.scatter(results[0,:],results[1,:],c=results[2,:], cmap='YlGnBu',marker='o',s=10, alpha=0.3)
    plt.colorbar()
    plt.scatter(sdp_m,rp_m,marker='*',color='r',s=100, label='Maximum Sharpe Ratio by Gradient descent method')    
    plt.scatter(minv, minvr,marker='*',color='b',s=100, label='Global Minimum Variance Portfolio')  
    plt.scatter(market_vol,market_ret,marker='d',color='g',s=100, label='Market portfolio') 
    plt.plot(risk, given_r, linestyle='-.', color='black', label='efficient frontier')
    plt.title('Optimal Portfolio Stimulation Method Comparision')
    plt.xlabel('Volatility')
    plt.ylabel('Returns %')
    plt.legend(labelspacing=0.8)   
```

```python
#############################################Results analysis by using Monte Carlo Method######################################
import statistics  

Result_data = pd.read_csv('F:\Python codes\Product model portfolio selection\Result\Result_rp_sdp.csv')
Result_data = Result_data.drop(Result_data.columns[[0]], axis=1)   #Drop the 1st col
RD = Result_data.values.tolist()
p_return=RD[0]
p_var=RD[1]
mean_p_return = sum(p_return)/ len(p_return)
var_p_return = statistics.variance(p_return)
print('The mean return of the portfolio after running',n, 'times is',round(mean_p_return,5),'%',', and the variance of the return is', round(var_p_return,5))
print('With 95% of confidence interval, the portfolio return is:',mean_p_return,'±', 1.96*var_p_return/math.sqrt(n))

plt.subplot(211)
plt.plot(range(1,n+1),p_return,marker='*',color='r')
plt.title('Optimal Portfolio Comparision')
plt.xlabel('Number of times running')
plt.xticks(np.arange(1,n+1,(n/10)-1))
plt.ylabel('Portfolio returns (%)')

plt.subplot(212)
plt.plot(range(1,n+1),p_var,marker='*',color='b')
plt.xlabel('Number of times running')
plt.xticks(np.arange(1,n+1,(n/10)-1))
plt.ylabel('Portfolio variance (%)')

sharpe_list = []
for i in range (n):
    sharpe_list.append(p_return[i]/(p_var[i]*100))
index_max_sharpe = sharpe_list.index(max(sharpe_list)) #count from 0

Weight_data = pd.read_csv('F:\Python codes\Product model portfolio selection\Result\Result_weight.csv')
Weight_data = Weight_data.drop(Weight_data.columns[[0]], axis=1)   #Drop the 1st col
print('The optimal portfolio weightage allocation with maximum sharpe ratio in',n,'trail by using Monte Carlo Method is: \n',Weight_data[Weight_data.columns[index_max_sharpe]])
print('The maximum sharpe ratio is:', max(sharpe_list))

############## M^2 risk adjusted performance############
M_square = max(sharpe_list) * market_vol*100 + rf
M_alpha = M_square - market_ret
print('The α of risk-adjusted perfomance of the optimal portfoilo is', M_alpha)
```
